```python
import itertools
import matplotlib.pyplot as plt
import numpy as np
import collections

elementos = [1,2,3,4,5,6,7,8,9]

entradas = list(map(list,itertools.permutations(elementos)))

def bubbleSort(alist):
    for passnum in range(len(alist)-1,0,-1):
        count = 0
        for i in range(passnum):
            if alist[i]>alist[i+1]:
                temp = alist[i]
                alist[i] = alist[i+1]
                alist[i+1] = temp
                count = count+2
            else:
                count = count+1
        return count

comparaciones=[]
for i in range(0, len(entradas)):
    bubbleSort(entradas[i])
    comparaciones.append(bubbleSort(entradas[i]))
print(comparaciones)
```

```python
counter = collections.Counter(comparaciones)
#print(counter)
#print(counter.values())
#print(counter.keys())
width = 1/1.5
plt.bar(counter.keys(), counter.values(), width, color="green")
```
